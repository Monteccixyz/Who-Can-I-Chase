```python
import pandas as pd

import os

weather_df = pd.read_csv('../data/processed/weather_andalucia_2025.csv')
bird_df = pd.read_csv(
    '../data/raw/ebird_spain_2020-2025.txt',
                      delimiter='\t',
                      usecols=[
                          'COMMON NAME',
                          'SCIENTIFIC NAME',
                          'OBSERVATION DATE',
                          'LATITUDE',
                          'LONGITUDE',
                          'STATE'
                      ])

```

```python
# Filter to Andalucía
bird_df = bird_df[bird_df['STATE'] == 'Andalucía']

# Round coordinates to match weather data
bird_df['lat_rounded'] = bird_df['LATITUDE'].round(1)
bird_df['lon_rounded'] = bird_df['LONGITUDE'].round(1)

# Rename date column to match
bird_df['date'] = bird_df['OBSERVATION DATE']
```

```python
# Filter to 2025 only
bird_df = bird_df[bird_df['date'].str.startswith('2025')]

# Strip the time part from weather dates to match
weather_df['date'] = weather_df['date'].str[:10]

print(weather_df['date'].head())
print(bird_df['date'].head())
```

```python
weather_df['latitude'] = weather_df['latitude'].round(1)
weather_df['longitude'] = weather_df['longitude'].round(1)
```

```python
merged = pd.merge(bird_df, weather_df,
                  left_on=['date', 'lat_rounded', 'lon_rounded'],
                  right_on=['date', 'latitude', 'longitude'])

print(len(merged))
```

```python
print(merged.head())
print(merged.columns.tolist())
```

```python
merged = merged.drop(columns=['latitude', 'longitude', 'OBSERVATION DATE'])
```

```python
merged.to_csv('../data/processed/training_data.csv', index=False)
print(f"Saved {len(merged)} rows")
```
